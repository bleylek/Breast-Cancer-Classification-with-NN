```python
import numpy as np
import pandas as pd
import sklearn.datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
bc_dataset=sklearn.datasets.load_breast_cancer()
```

```python
dataFrame['label'] = bc_dataset.target
```

```python
X = dataFrame.drop(columns='label',axis=1)
y = dataFrame['label']
```

```python
print(X)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

Standardizing the Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

Building the NN

```python
import tensorflow as tf
tf.random.set_seed(3) # as random_state
from tensorflow import keras
```

Setting up the layers of NN

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)), # one input layer (Flatten is for converting our data into a single dimensional array)
    keras.layers.Dense(20, activation ='relu'), # one hidden layer, 20 neurons
    keras.layers.Dense(2, activation='sigmoid') # one output layer, 2 neurons (we are creating two neurons in the output layer which is equal to the number of classes that we have (0 and 1))
])
```

Compiling the NN

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy', # ex/ dog-0,cat-1,horse-2
              metrics=['accuracy'])
```

Training the NN

```python
history = model.fit(X_train, y_train, validation_split=0.1, epochs=10) # without standardization
```

```python
history = model.fit(X_train_std, y_train, validation_split=0.1, epochs=10)
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc='upper right')
```

Accuracy of the Model on Test Data

```python
loss, accuracy = model.evaluate(X_test_std, y_test)
print(accuracy)
```
